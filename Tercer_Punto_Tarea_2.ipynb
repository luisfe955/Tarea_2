```python
#Lanzamiento de dardos en una dimensión.

# El experimento consiste en lanzar 100(k_lanzamientos) veces un dardo y repetir 
# 1000(k_repeticiones) veces los 100 lanzamientos. Este experimeto se repite 
# 200(k_experimentos) veces.

# La hipótesis a verificar es que la media es cero.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Lanzamiento de dardos una dimensión.
k_lanzamientos=100 # cantidad de disparos que se van a generar por repetición. 
k_repeticiones=1000 # cantidad de repeticiones por experimento. 
k_experimentos=200 # cantidad de experimentos

anchox = 1. # ancho de la distribución normal 1 en x.    
centrox = 0 # x centro de la distribución los disparos (hipótesis).

p_value=[] # Arreglo para guardar los valores de p value de cada experimento. 
N_experiments=[] # Número de veces que se va a repetir el experimento {1,2,...,200}. 

# ciclo que repite el experimento 200 veces.
for i in range(k_experimentos):
    N_experiments.append(i+1)
    x_media=[]
    x_mediatotal=0.
    for j in range(k_repeticiones):
        media=0.
        x_suma=0.
        for k in range(k_lanzamientos):
            x = np.random.normal(centrox,anchox) # distribución normal de los x.
            #print x
            x_suma+=x
        media=(x_suma)/k_lanzamientos    
        x_media.append(media)
        x_mediatotal+=media
    desv_es=np.std(x_media)
    x_mediatotal=x_mediatotal/k_repeticiones
    t_student=(x_mediatotal-centrox)/desv_es
    p=1-stats.t.sf(np.abs(t_student), 1)*2
    p_value.append(p)
plt.scatter(N_experiments,p_value)
plt.xlabel ('N_experiments')
plt.ylabel ('1-P_value')
```

```python
# Para la mayoria de los experimentos la hipótesis es verdadera ya que 1-p_value<0.05. 
# Sin embargo es posible observar que existen pocos exprimentos en los cuales la hipótesis 
# es rechazada.
```
